```python
from imblearn.under_sampling import RandomUnderSampler

# 각 클래스의 샘플 수 확인
class_counts = y.value_counts()

# 다수 클래스와 소수 클래스 확인
majority_class = class_counts.idxmax()
minority_class = class_counts.idxmin()

# 2:1 비율로 다운샘플링
sampling_strategy = {majority_class: class_counts[minority_class] * 2}

# 다운샘플링 객체 생성
rus = RandomUnderSampler(sampling_strategy=sampling_strategy, random_state=42)

# 다운샘플링 적용
X_resampled, y_resampled = rus.fit_resample(X, y)

# 결과 확인
print(f"다운샘플링 후 클래스 비율:\n{pd.Series(y_resampled).value_counts()}")
```

```python
# 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=21)
```

```python
from sklearn.model_selection import KFold
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import accuracy_score
import numpy as np

# 업데이트된 categorical_columns 리스트 (범주형 변수들)
categorical_columns = [col for col in X_resampled.columns if col in categorical_df.columns]

# CatBoostClassifier 모델 초기화 (최적화된 파라미터 설정)
model = CatBoostClassifier(
    iterations=1000,             # 반복 횟수
    learning_rate=0.05,          # 학습률 (낮춤)
    depth=4,                     # 트리 깊이 (낮춤)
    l2_leaf_reg=10,              # L2 정규화 계수 (증가)
    bagging_temperature=1.5,     # 배깅 온도 (증가)
    subsample=0.8,               # 학습 데이터의 80%만 사용
    loss_function='Logloss',     # 손실 함수
    cat_features=categorical_columns,  # 범주형 피처 지정
    verbose=100                  # 학습 과정 출력
)

# K-Fold 교차 검증 설정 (10-폴드 교차 검증)
kf = KFold(n_splits=10, shuffle=True, random_state=21)

# 교차 검증 점수 저장할 리스트
cv_scores = []

# K-Fold 교차 검증 수행
for train_index, val_index in kf.split(X_resampled):
    # 훈련 및 검증 데이터 분할
    X_train, X_val = X_resampled.iloc[train_index], X_resampled.iloc[val_index]
    y_train, y_val = y_resampled.iloc[train_index], y_resampled.iloc[val_index]
    
    # 데이터 Pool 생성 (CatBoost 전용)
    train_pool = Pool(X_train, y_train, cat_features=categorical_columns)
    val_pool = Pool(X_val, y_val, cat_features=categorical_columns)
    
    # 모델 학습
    model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=30, verbose=False)
    
    # 검증 데이터에 대한 예측
    y_pred = model.predict(X_val)
    
    # 정확도 계산 및 저장
    accuracy = accuracy_score(y_val, y_pred)
    cv_scores.append(accuracy)

# 교차 검증 평균 성능 출력
mean_accuracy = np.mean(cv_scores)
print(f"Mean Validation Accuracy: {mean_accuracy:.4f}")
```
